## 1.1 Feature Selection <a class="anchor" id="11"></a>

### 1.1.1. Importing Libraries <a class="anchor" id="111"></a>

```python
import pandas as pd 
import numpy as np  
import seaborn as sns
import matplotlib.pyplot as plt
```

### 1.1.2. Converting categorical data to numerical <a class="anchor" id="112"></a>

```python
df = pd.read_csv('data/training_data.csv')
```

```python
df['company']=df['company'].astype('category').cat.codes
df['country']=df['country'].astype('category').cat.codes
df['director']=df['director'].astype('category').cat.codes
df['genre']=df['genre'].astype('category').cat.codes
df['rating']=df['rating'].astype('category').cat.codes
df['star']=df['star'].astype('category').cat.codes
df['writer']=df['writer'].astype('category').cat.codes
df['adult']=df['adult'].astype('category').cat.codes
```

### 1.1.3. Correlation Matrix <a class="anchor" id="113"></a>

```python
corr_df=df[['budget_norm','company','country','director','genre','gross','rating','runtime_norm','score_norm','star','votes_norm',
            'year','popularity_norm','year_released','profitability_ratio']].corr()
```

```python
corr_df
```

### 1.1.4. Heat-Map of correlation Matrix <a class="anchor" id="114"></a>

```python
plt.figure(figsize=(20,15))
sns.heatmap(corr_df,cmap='coolwarm',vmin=-1, vmax=1,annot=True,linewidth=.5);
```

### 1.1.5. Relevant attributes to improve the accuracy of data mining results <a class="anchor" id="115"></a>

```python
corr_df_features = set()
for i in range(len(corr_df.columns)):
    if corr_df.iloc[i,14]<-0.1 or corr_df.iloc[i,14]>0.1:
        corr_df_features.add(corr_df.columns[i])
```

```python
final_features = [e for e in corr_df_features if e not in ['gross','profitability_ratio']]
```
